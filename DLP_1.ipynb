```python
# Install necessary libraries
!pip install datasets pandas

from datasets import load_dataset, load_from_disk
from google.colab import files
import pandas as pd
import os

# Load IMDB dataset
imdb_dataset = load_dataset("stanfordnlp/imdb")
print(imdb_dataset)

# Train-test split
train_dataset = imdb_dataset['train']
small_ds = train_dataset.train_test_split(test_size=0.3)
print(small_ds)

# Upload and process CSV file
uploaded = files.upload()
file_name = 'annual-enterprise-survey-2023-financial-year-provisional-size-bands.csv'

try:
    df = pd.read_csv(file_name)
    print(df.head())
except FileNotFoundError:
    print(f"{file_name} not found. Please upload the correct file.")

# Load the CSV file into Hugging Face Dataset
dataset = load_dataset('csv', data_files=file_name)

# Save and reload in Arrow format
dataset.save_to_disk('arrow_dataset')
arrow_dataset = load_from_disk('arrow_dataset')
print(arrow_dataset)
```
